```python
filenames=os.listdir('/content/img_resized')
label=[]
for i in range(2000):
  file_name=filenames[i]
  name=file_name[0:3]
  if(name=='dog'):
    label.append(1)
  else:
    label.append(0)
```

```python
import cv2
import glob
```

```python
image_directory = '/content/img_resized/'
image_extension = ['png', 'jpg']

files = []

[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]

dog_cat_images = np.asarray([cv2.imread(file) for file in files])
```

```python
print(dog_cat_images)
```

```python
dog_cat_images.shape
```

```python
type(dog_cat_images)
```

```python
X=dog_cat_images
Y=np.asarray(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=24)
```

```python
X_train.shape,Y_train.shape,X_test.shape,Y_test.shape
```

```python
X_train_std=X_train/255
X_test_std=X_test/255
```

```python
print(X_train_std)
```

```python
print(X_test_std)
```

```python
import tensorflow as tf
import tensorflow_hub as hub

mobilenet_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/4"

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224, 224, 3), trainable=False)
```

```python
from tensorflow import keras
```

```python
model=keras.Sequential([
    pretrained_model,
    keras.layers.Dense(2)
])
model.summary()
```

```python
model.compile( optimizer='adam',
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy']
               )
```

```python
model.fit(X_train_std,Y_train,epochs=5)
```

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print('Loss:',loss)
print('accuracy:',accuracy)
```
